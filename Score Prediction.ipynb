 ___The aim of this project is to predict the scores of students based on the number of hours they studied for, using simple Linear Regression model___

```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Loading the data into the dataframe df
link = "http://bit.ly/w-data"
df = pd.read_csv(link)
df.head()
```

```python
# number of rows and columns in the dataset
df.shape
```

```python
# Plotting a scatterplot with regression line
sns.jointplot(x="Hours", y="Scores",data = df,kind="reg",joint_kws={'line_kws':{'color':'cyan'}})
```

___Inference___

- _From the graph, it is understood that Hours studied and Marks scored are in a positive linear relationship._

```python
# Creating X 2-D array object to store the Hours variable 
X = df[['Hours']]
```

```python
# creating a y object to store the dependent variable - Scores
y = df['Scores']
```

## Test-train split

```python
from sklearn.model_selection import train_test_split
```

```python
# performing test-train split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size =0.2,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# creating a linear regression model
model = LinearRegression()
```

```python
# fitting the data into the model
model.fit(X_train,y_train)
```

```python
# printing the intercept value
print(model.intercept_)
```

```python
# printing the coefficient value
print(model.coef_)
```

## Predicting scores

```python
y_pred = model.predict(X_test)
```

```python
# printing actual scores and predicted scores
df2 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df2
```

## Model Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error,mean_absolute_error
```

```python
print("The r2_score of predicted model is: ",r2_score(y_test,y_pred))
```

```python
print("The mean absolute error is: ",mean_absolute_error(y_test,y_pred))
```

```python
print("The mean squared error is: ",mean_squared_error(y_test,y_pred))
```
